### Çok formatlı giriş

```python
#Secret al
from google.colab import userdata
GEMINI_API=userdata.get('GEMINI_API')
```

```python
#gen ai kütüphanesini içe aktar
from google import genai
```

```python
#istemci oluştur
client=genai.Client(api_key=GEMINI_API)
```

```python
#Modeli seç
MODEL='gemini-3.1-flash-lite'
```

```python
#yükelyeceğimizi görseli görmek için
from PIL import Image
img = Image.open("/content/piyano.jpg")
display(img)
```

```python
#client a dosya yükelmek için
uploaded_file=client.files.upload(file='/content/piyano.jpg')
```

```python
# AI etkileşim
interaction=client.interactions.create(
    model=MODEL,
    input=[
        {"type":"text",
         "text":"Görüntüdeki nesneyi açıkla."},
        {"type":"image",
         "uri":uploaded_file.uri,
         "mime_type":uploaded_file.mime_type
        }

    ]
)
```

```python
print(interaction.output_text)
```
