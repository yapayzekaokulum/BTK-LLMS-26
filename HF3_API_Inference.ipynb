### Test

```python
from huggingface_hub import InferenceClient
client = InferenceClient(token=HF_TOKEN)

r = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[{"role": "user", "content": "Türkiye'nin başkenti neresidir?"}],
    max_tokens=256,          # ← 10 değil; reasoning + cevaba yer aç
)
msg = r.choices[0].message
# reasoning modelinde content bazen None; reasoning alanına da bak
print(msg.content or getattr(msg, "reasoning_content", None) or getattr(msg, "reasoning", msg))
```
