```python
from google import genai
```

```python
client=genai.Client(api_key=GEMINI_API)
```

```python
# modellerin adını listele
for model in client.models.list():
    print(model.name)
```

```python
MODEL='gemini-3.1-flash-lite'
# Yeni kullanıcılara bu model ücretsiz planda sunulmuyor.
```

```python
INPUT=input("Sorunuzu yazınız!")
```

```python
interaction=client.interactions.create(
model=MODEL,input=INPUT
)
```

```python
print(interaction.output_text) # yanıtı çıktıyı veriyor
```
